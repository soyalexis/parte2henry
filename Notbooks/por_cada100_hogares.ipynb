```python
# Reemplazar comas por puntos y convertir a tipo float
df['Accesos por cada 100 hogares'] = df['Accesos por cada 100 hogares'].str.replace(',', '.').astype(float)
```

```python
# Gráfico de evolución para cada provincia
provincias = df['Provincia'].unique()

plt.figure(figsize=(15, 8))

for provincia in provincias:
    data_provincia = df[df['Provincia'] == provincia]
    plt.plot(data_provincia['Año'], data_provincia['Accesos por cada 100 hogares'], label=provincia)

plt.xlabel('Año')
plt.ylabel('Accesos por cada 100 hogares (Redondeado a 2 decimales)')
plt.title('Evolución de Accesos por cada 100 hogares en cada Provincia (2014-2022)')
plt.legend(loc='upper left', bbox_to_anchor=(1,1))

plt.show()
```

En general podemos ver que hay una tendencia positiva de acceso a internet en todas las provincias.

```python
# Agrupamos por año y trimestre. Y calculamos la media
df_agrupado_trimestral = df.groupby(['Año', 'Trimestre'])['Accesos por cada 100 hogares'].mean().reset_index()

# Grafico de tendencias trimestrales
plt.figure(figsize=(12, 6))

for año in df_agrupado_trimestral['Año'].unique():
    df_año = df_agrupado_trimestral[df_agrupado_trimestral['Año'] == año]
    plt.plot(df_año['Trimestre'], df_año['Accesos por cada 100 hogares'], label=str(año))

plt.xlabel('Trimestre')
plt.ylabel('Media de Accesos por cada 100 hogares')
plt.title('Tendencias Trimestrales de Accesos a Internet (2014-2022)')
plt.legend()
plt.show()
```

A través de este gráfico, se destaca un pico significativo en el acceso a Internet durante el segundo y tercer trimestre de 2017, indicando una demanda particularmente alta en ese período. Posteriormente, la estabilidad se mantuvo para el cuarto trimestre de ese año.

En contraste, observamos una marcada disminución en los accesos durante 2019, especialmente después del tercer trimestre. Este declive puede sugerir cambios en la demanda o factores externos que afectaron la conectividad en ese período, como por ejemplo el Covid.

Una tendencia positiva se manifiesta en el año 2022, donde los accesos muestran un incremento constante. Este aumento podría estar vinculado a mejoras en la infraestructura, cambios en la demanda del mercado o la implementación de nuevas tecnologías.
