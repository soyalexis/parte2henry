```python
plt.figure(figsize=(12, 8))

# Gr叩fico de barras para Ingresos por trimestre
sns.barplot(x='Trimestre', y='Ingresos (miles de pesos)', hue='A単o', data=df)

plt.title('Ingresos por Trimestre')
plt.xlabel('Trimestre')
plt.ylabel('Ingresos (miles de pesos)')
plt.legend(title='A単o')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 8))

# Gr叩fico de barras para Ingresos por a単o
sns.barplot(x='A単o', y='Ingresos (miles de pesos)', data=df)

plt.title('Ingresos por A単o')
plt.xlabel('A単o')
plt.ylabel('Ingresos (miles de pesos)')
plt.grid(True)
plt.show()
```

Claramente se ve un aumento significativo en los ingresos a単o tras a単o pero al estar en pesos no tiene en cuenta la inflacion lo cual modifica mucho el analisis que se pueda hacer en base al aumento de ingresos
