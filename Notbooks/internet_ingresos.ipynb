```python
plt.figure(figsize=(12, 8))

# Gráfico de barras para Ingresos por trimestre
sns.barplot(x='Trimestre', y='Ingresos (miles de pesos)', hue='Año', data=df)

plt.title('Ingresos por Trimestre')
plt.xlabel('Trimestre')
plt.ylabel('Ingresos (miles de pesos)')
plt.legend(title='Año')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 8))

# Gráfico de barras para Ingresos por año
sns.barplot(x='Año', y='Ingresos (miles de pesos)', data=df)

plt.title('Ingresos por Año')
plt.xlabel('Año')
plt.ylabel('Ingresos (miles de pesos)')
plt.grid(True)
plt.show()
```

Claramente se ve un aumento significativo en los ingresos año tras año pero al estar en pesos no tiene en cuenta la inflacion lo cual modifica mucho el analisis que se pueda hacer en base al aumento de ingresos
